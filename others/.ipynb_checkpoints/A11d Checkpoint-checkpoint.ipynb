```python
# Add Earth Engine dataset
image = ee.Image("USGS/SRTMGL1_003")

#  ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
#  Chapter:      A1.1 Agricultural Environments
#  Checkpoint:   A11d
#  Authors:      Sherrie Wang, George Azzari
#  ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

##############################
# 1. Pull all Landsat 7 and 8 images over study area
##############################

# Define study area.
TIGER = ee.FeatureCollection('TIGER/2018/Counties')
region = ee.Feature(TIGER \
    .filter(ee.Filter.eq('STATEFP', '17')) \
    .filter(ee.Filter.eq('NAME', 'McLean')) \
    .first())
geometry = region.geometry()
Map.centerObject(region)
Map.addLayer(region, {
    'color': 'red'
}, 'McLean County')

# Import Landsat imagery.
landsat7 = ee.ImageCollection('LANDSAT/LE07/C02/T1_L2')
landsat8 = ee.ImageCollection('LANDSAT/LC08/C02/T1_L2')

# Functions to rename Landsat 7 and 8 images.
def renameL7(img):
    return img.rename(['BLUE', 'GREEN', 'RED', 'NIR', 'SWIR1',
        'SWIR2', 'TEMP1', 'ATMOS_OPACITY', 'QA_CLOUD',
        'ATRAN', 'CDIST',
        'DRAD', 'EMIS', 'EMSD', 'QA', 'TRAD', 'URAD',
        'QA_PIXEL',
        'QA_RADSAT'
    ])


def renameL8(img):
    return img.rename(['AEROS', 'BLUE', 'GREEN', 'RED', 'NIR',
        'SWIR1',
        'SWIR2', 'TEMP1', 'QA_AEROSOL', 'ATRAN', 'CDIST',
        'DRAD', 'EMIS',
        'EMSD', 'QA', 'TRAD', 'URAD', 'QA_PIXEL', 'QA_RADSAT'
    ])


# Functions to mask out clouds, shadows, and other unwanted features.
def addMask(img):
    # Bit 0: Fill
    # Bit 1: Dilated Cloud
    # Bit 2: Cirrus (high confidence) (L8) or unused (L7)
    # Bit 3: Cloud
    # Bit 4: Cloud Shadow
    # Bit 5: Snow
    # Bit 6: Clear
    #        0: Cloud or Dilated Cloud bits are set
    #        1: Cloud and Dilated Cloud bits are not set
    # Bit 7: Water
    clear = img.select('QA_PIXEL').bitwiseAnd(64).neq(0)
    clear = clear.updateMask(clear).rename(['pxqa_clear'])

    water = img.select('QA_PIXEL').bitwiseAnd(128).neq(0)
    water = water.updateMask(water).rename(['pxqa_water'])

    cloud_shadow = img.select('QA_PIXEL').bitwiseAnd(16).neq(0)
    cloud_shadow = cloud_shadow.updateMask(cloud_shadow).rename([
        'pxqa_cloudshadow'
    ])

    snow = img.select('QA_PIXEL').bitwiseAnd(32).neq(0)
    snow = snow.updateMask(snow).rename(['pxqa_snow'])

    masks = ee.Image.cat([
        clear, water, cloud_shadow, snow
    ])

    return img.addBands(masks)


def maskQAClear(img):
    return img.updateMask(img.select('pxqa_clear'))


# Function to add GCVI as a band.
def addVIs(img):
    gcvi = img.expression('(nir / green) - 1', {
      'nir': img.select('NIR'),
      'green': img.select('GREEN')
    }).select([0], ['GCVI'])

    return ee.Image.cat([img, gcvi])


# Define study time period.
start_date = '2020-01-01'
end_date = '2020-12-31'

# Pull Landsat 7 and 8 imagery over the study area between start and end dates.
landsat7coll = landsat7 \
    .filterBounds(geometry) \
    .filterDate(start_date, end_date) \
    .map(renameL7)

landsat8coll = landsat8 \
    .filterDate(start_date, end_date) \
    .filterBounds(geometry) \
    .map(renameL8)

# Merge Landsat 7 and 8 collections.
landsat = landsat7coll.merge(landsat8coll) \
    .sort('system:time_start')

# Mask out non-clear pixels, add VIs and time variables.
landsat = landsat.map(addMask) \
    .map(maskQAClear) \
    .map(addVIs)

```

```python
# Define la funcin para obtener los valores de GCVI en el punto de inters
def get_gcvi_values(image):
    value = image.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=point,
        scale=30
    ).get('GCVI')
    date = image.date().format("YYYY-MM-dd")
    return ee.Feature(None, {'GCVI': value, 'date': date})

# Obtener los valores de GCVI como una coleccin de features
gcvi_series = landsat.map(get_gcvi_values).filter(ee.Filter.notNull(['GCVI']))

# Convertir la coleccin a una lista de diccionarios en Python
gcvi_list = gcvi_series.getInfo()['features']
gcvi_data = [{'date': item['properties']['date'], 'GCVI': item['properties']['GCVI']} for item in gcvi_list]

# Convertir a un DataFrame de pandas
df = pd.DataFrame(gcvi_data)

# Convertir la columna de fechas a tipo datetime
df['date'] = pd.to_datetime(df['date'])

# Plotear la serie de tiempo de GCVI usando matplotlib
plt.figure(figsize=(10, 5))
plt.plot(df['date'], df['GCVI'], marker='o', linestyle='-', color='b')
plt.title('Landsat GCVI time series')
plt.xlabel('Date')
plt.ylabel('GCVI')
plt.grid(True)
plt.show()
```

```python
# Function that adds time band to an image.
def addTimeUnit(image, refdate):
    date = image.date()

    dyear = date.difference(refdate, 'year')
    t = image.select(0).multiply(0).add(dyear).select([0], ['t']) \
        .float()

    imageplus = image.addBands(t)

    return imageplus


# Function that adds harmonic basis to an image.
def addHarmonics(image, omega, refdate):
    image = addTimeUnit(image, refdate)
    timeRadians = image.select('t').multiply(2 * math.pi * omega)
    timeRadians2 = image.select('t').multiply(4 * math.pi *
    omega)

    return image \
        .addBands(timeRadians.cos().rename('cos')) \
        .addBands(timeRadians.sin().rename('sin')) \
        .addBands(timeRadians2.cos().rename('cos2')) \
        .addBands(timeRadians2.sin().rename('sin2')) \
        .addBands(timeRadians.divide(timeRadians) \
            .rename('constant'))


# Apply addHarmonics to Landsat image collection.
omega = 1
# Definir la funcin addHarmonics por separado
def apply_harmonics(image):
    return addHarmonics(image, omega, start_date)

# Aplicar la funcin addHarmonics a la coleccin de imgenes usando map
landsatPlus = landsat.map(apply_harmonics)

print('Landsat collection with harmonic basis: ', landsatPlus)

```

```python
# Function to run linear regression on an image.
def arrayimgHarmonicRegr(harmonicColl, dependent, independents):
    independents = ee.List(independents)
    dependent = ee.String(dependent)

    regression = harmonicColl \
        .select(independents.add(dependent)) \
        .reduce(ee.Reducer.linearRegression(independents.length(), 1))

    return regression

# Function to extract and rename regression coefficients.
def imageHarmonicRegr(harmonicColl, dependent, independents):
    hregr = arrayimgHarmonicRegr(harmonicColl, dependent, independents)

    independents = ee.List(independents)
    dependent = ee.String(dependent)

    def func_mxk(b):
        return dependent.cat(ee.String('_')).cat(ee.String(b))

    newNames = independents.map(func_mxk)

    imgCoeffs = hregr.select('coefficients') \
        .arrayProject([0]) \
        .arrayFlatten([independents]) \
        .select(independents, newNames)

    return imgCoeffs

# Function to apply imageHarmonicRegr and create a multi-band image.
def apply_imageHarmonicRegr(band):
    return imageHarmonicRegr(harmonicColl, band, independents)

# Apply getHarmonicCoeffs to ImageCollection.
def getHarmonicCoeffs(harmonicColl, bands, independents):
    coefficients = ee.ImageCollection.fromImages(
        bands.map(apply_imageHarmonicRegr)
    )
    return coefficients.toBands()
```

```python
# Definir harmonicColl antes de usarla, si no est definida antes en tu cdigo
harmonicColl = landsatPlus

# Aplicar la funcin corregida
bands = ee.List(['NIR', 'SWIR1', 'SWIR2', 'GCVI'])  # Cambi la lista a un objeto ee.List
independents = ee.List(['constant', 'cos', 'sin', 'cos2', 'sin2'])
harmonics = getHarmonicCoeffs(harmonicColl, bands, independents)

harmonics = harmonics.clip(geometry)
harmonics = harmonics.multiply(10000).toInt32()

# Compute fitted values.
gcviHarmonicCoefficients = harmonics \
    .select([
        '3_GCVI_constant', '3_GCVI_cos',
        '3_GCVI_sin', '3_GCVI_cos2', '3_GCVI_sin2'
    ]) \
    .divide(10000)

def add_fitted_band(image):
    return image.addBands(
        image.select(independents) \
        .multiply(gcviHarmonicCoefficients) \
        .reduce('sum') \
        .rename('fitted')
    )

fittedHarmonic = landsatPlus.map(add_fitted_band)
```

```python
# Parmetros de resampling temporal
start_date = '2020-01-01'
end_date = '2020-12-31'
date_range = ee.DateRange(start_date, end_date)

# Definir la funcin para calcular el valor ajustado de GCVI en un da especfico
def harmonic_fit(date, harmonic_coefficients, independents):
    date = ee.Date(date)
    time = date.difference(ee.Date(start_date), 'year')
    
    omega = 1
    timeRadians = ee.Image.constant(time).multiply(2 * math.pi * omega)
    timeRadians2 = ee.Image.constant(time).multiply(4 * math.pi * omega)
    
    # Crear las bandas de senos y cosenos
    cos = timeRadians.cos().rename('cos')
    sin = timeRadians.sin().rename('sin')
    cos2 = timeRadians2.cos().rename('cos2')
    sin2 = timeRadians2.sin().rename('sin2')
    constant = ee.Image.constant(1).rename('constant')
    
    # Combinar las bandas
    harmonics = ee.Image.cat([constant, cos, sin, cos2, sin2])
    
    # Calcular el valor ajustado usando los coeficientes
    fitted = harmonics.multiply(harmonic_coefficients).reduce('sum').rename('fitted_gcvi')
    
    return fitted.set('system:time_start', date.millis())

# Crear una lista de fechas diarias
dates = ee.List.sequence(ee.Date(start_date).millis(), ee.Date(end_date).millis(), 24 * 60 * 60 * 1000)

# Mapear la funcin sobre la lista de fechas para generar una coleccin de imgenes diarias
daily_harmonics = ee.ImageCollection.fromImages(
    dates.map(lambda millis: harmonic_fit(ee.Date(millis), gcviHarmonicCoefficients, independents))
)
```

```python
#daily_harmonics.getInfo()

def addDate(image):
    img_date = ee.Date(image.date())
    img_date = ee.Number.parse(img_date.format('YYYYMMdd'))
    return image.addBands(ee.Image(img_date).rename('date').toInt())

def addMonth(image):
    img_date = ee.Date(image.date())
    img_doy = ee.Number.parse(img_date.format('M'))
    return image.addBands(ee.Image(img_doy).rename('month').toInt())

def addDOY(image):
    img_date = ee.Date(image.date())
    img_doy = ee.Number.parse(img_date.format('D'))
    return image.addBands(ee.Image(img_doy).rename('doy').toInt())


def invertCloudProbabilityBand(img):
    cldProb = img.select('fitted_gcvi')
    cldProbInv = cldProb.multiply(-1).rename('quality')
    return img.addBands(cldProbInv)


greenest = daily_harmonics.map(addDate).map(addMonth).map(addDOY).qualityMosaic('fitted_gcvi').select('fitted_gcvi')
```

```python
region = ee.FeatureCollection('TIGER/2018/Counties') \
    .filter(ee.Filter.eq('STATEFP', '17')) \
    .filter(ee.Filter.eq('NAME', 'McLean')) \
    .geometry()

# Funcin para calcular RMSE y R usando la correlacin manualmente
def calculate_metrics(img):
    # Extraer la banda GCVI de landsat
    gcvi = img.select('GCVI')
    
    # Extraer la banda fitted_gcvi de fittedHarmonic
    fitted_gcvi = img.select('fitted')
    
    # Calcular RMSE
    residual = gcvi.subtract(fitted_gcvi)
    squared_residual = residual.pow(2)
    rmse = squared_residual.reduce(ee.Reducer.mean()).sqrt().rename('RMSE')
    
    # Calcular las medias de GCVI y fitted_gcvi
    mean_gcvi = gcvi.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=region,
        scale=30,
        bestEffort=True
    ).get('GCVI')
    
    mean_fitted_gcvi = fitted_gcvi.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=region,
        scale=30,
        bestEffort=True
    ).get('fitted')
    
    # Convertir los valores medios a imgenes constantes
    mean_gcvi_img = ee.Image.constant(mean_gcvi)
    mean_fitted_gcvi_img = ee.Image.constant(mean_fitted_gcvi)
    
    # Calcular la covarianza
    covariance = gcvi.subtract(mean_gcvi_img).multiply(fitted_gcvi.subtract(mean_fitted_gcvi_img)).reduce(ee.Reducer.mean())
    
    # Calcular las desviaciones estndar
    std_dev_gcvi = gcvi.subtract(mean_gcvi_img).pow(2).reduce(ee.Reducer.mean()).sqrt()
    std_dev_fitted_gcvi = fitted_gcvi.subtract(mean_fitted_gcvi_img).pow(2).reduce(ee.Reducer.mean()).sqrt()
    
    # Calcular la correlacin
    correlation = covariance.divide(std_dev_gcvi.multiply(std_dev_fitted_gcvi))
    
    # Calcular R como el cuadrado de la correlacin
    r2 = correlation.pow(2).rename('R2')
    
    # Asegurar que R est en el rango [0, 1]
    r2 = r2.max(0).min(1)
    
    return img.addBands(rmse).addBands(r2)

# Combinar las colecciones usando la funcin .map()
combined = landsat.map(lambda img: img.addBands(fittedHarmonic.filterDate(img.date()).first().select('fitted')))

# Aplicar la funcin de clculo de mtricas a la coleccin combinada
metrics_images = combined.map(calculate_metrics)

# Reducir la coleccin de imgenes para obtener las medias de RMSE y R
rmse_image = metrics_images.select('RMSE').mean().clip(region)
r2_image = metrics_images.select('R2').mean().clip(region)

```

```python
import ee

# Inicializar Earth Engine
ee.Initialize()

# Definir la regin de inters
region = ee.FeatureCollection('TIGER/2018/Counties') \
    .filter(ee.Filter.eq('STATEFP', '17')) \
    .filter(ee.Filter.eq('NAME', 'McLean')) \
    .geometry()

# Funcin para calcular RMSE y R para una imagen individual
def calculate_metrics_for_image(img):
    independents = ee.List(['constant', 'cos', 'sin', 'cos2', 'sin2'])
    dependent = 'GCVI'
    
    # Calcular la regresin lineal
    regression = img.select(independents.add(dependent)) \
                    .reduce(ee.Reducer.linearRegression(independents.length(), 1))
    
    coefficients = regression.select('coefficients').arrayProject([0]).arrayFlatten([independents])
    fitted_values = coefficients.multiply(ee.Image.constant(ee.List.repeat(1, independents.length()))).reduce('sum').rename('fitted')
    
    # Calcular el RMSE
    residuals = img.select(dependent).subtract(fitted_values)
    squared_residuals = residuals.pow(2)
    rmse = squared_residuals.reduce(ee.Reducer.mean()).sqrt().rename(dependent + '_rmse')
    
    # Calcular la varianza del GCVI (variable dependiente)
    stats = img.select(dependent).reduce(ee.Reducer.sampleVariance().combine(ee.Reducer.count(), None, True).combine(ee.Reducer.mean(), None, True))
    variance = stats.select(dependent + '_variance')
    
    # Calcular el R
    r2 = ee.Image(1).subtract(rmse.pow(2).divide(variance)).rename(dependent + '_r2')
    
    return img.addBands(fitted_values).addBands(rmse).addBands(r2)

# Aplicar la funcin a cada imagen en la coleccin fittedHarmonic
metrics_images = fittedHarmonic.map(calculate_metrics_for_image)

# Reducir la coleccin de imgenes para obtener las medias de RMSE y R
rmse_image = metrics_images.select('GCVI_rmse').mean().clip(region)
r2_image = metrics_images.select('GCVI_r2').mean().clip(region)
```
